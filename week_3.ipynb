```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#reading the data
data_train=pd.read_csv('training_set_features.csv')
data_train_labels=pd.read_csv('training_set_labels.csv')
data_test=pd.read_csv('test_set_features.csv')
data_test_labels=pd.read_csv('submission_format.csv')
```

```python
data_train.nunique()
```

```python
data_train.isnull().sum()
```

```python
#dropping columns which are irrelevent or have high number of missing values
coln_to_drop=['respondent_id','health_insurance','employment_industry', 'employment_occupation']
data_train.drop(columns=coln_to_drop, inplace=True)
data_test.drop(columns=coln_to_drop, inplace=True)
data_train.isnull().sum()
```

```python
#filling missing values with zero for some columns
coln_to_zero=['xyz_concern', 'xyz_knowledge',
       'behavioral_antiviral_meds', 'behavioral_avoidance',
       'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_large_gatherings', 'behavioral_outside_home',
       'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal',
       'chronic_med_condition', 'child_under_6_months', 'health_worker']
data_train[coln_to_zero]=data_train[coln_to_zero].fillna(0)
data_test[coln_to_zero]=data_test[coln_to_zero].fillna(0)
data_train.isnull().sum()
```

```python
#filling missing values with mean for some columns
coln_mean=['opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults','household_children']
data_train[coln_mean]=data_train[coln_mean].fillna(data_train[coln_mean].mean()).astype(int)
data_test[coln_mean]=data_test[coln_mean].fillna(data_test[coln_mean].mean()).astype(int)
data_train.head()
```

```python
#filling missing values with mode for some columns
coln_mode=['income_poverty','education','marital_status','rent_or_own', 'employment_status']
data_train[coln_mode]=data_train[coln_mode].fillna(data_train[coln_mode].mode().iloc[0])
data_test[coln_mode]=data_test[coln_mode].fillna(data_test[coln_mode].mode().iloc[0])
```

```python
data_train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
coln_encode=['age_group','race','sex','hhs_geo_region','census_msa','employment_status','education','marital_status','rent_or_own','income_poverty']
data_train[coln_encode]=data_train[coln_encode].apply(le.fit_transform)
data_test[coln_encode]=data_test[coln_encode].apply(le.fit_transform)
data_train.head()
```

```python
#Below is code for logistic regression.
#This snippet works well on our dataset and has a better score.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train_xyz,x_test_xyz,y_train_xyz,y_test_xyz=train_test_split(data_train,data_train_labels['xyz_vaccine'],test_size=0.2,random_state=2)
x_train_ss,x_test_ss,y_train_ss,y_test_ss=train_test_split(data_train,data_train_labels['seasonal_vaccine'],test_size=0.2,random_state=2)
sam1=LogisticRegression()
sam2=LogisticRegression()
sam1.fit(x_train_xyz,y_train_xyz)
sam2.fit(x_train_ss,y_train_ss)
data_test_labels['xyz_vaccine']=sam1.predict_proba(data_test)[:,1]
data_test_labels['seasonal_vaccine']=sam2.predict_proba(data_test)[:,1]
sam1.score(x_train_xyz,y_train_xyz),sam2.score(x_train_ss,y_train_ss)
```
